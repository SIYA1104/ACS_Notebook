###### Discuss the shape of the distribution (skewness and Kurtosis) from the plot.

### Skewness: Skewness tells us the asymmetry of the data. If the skewness is:
-Positive: The tail on the right side is longer, and the distribution is right-skewed.<br>
-Negative: The tail on the left side is longer, and the distribution is left-skewed.<br>
-Zero: The distribution is symmetrical.

### Kurtosis: Kurtosis indicates the "tailedness" or the sharpness of the peak of the distribution.
-Positive Kurtosis: The data has heavy tails and a sharp peak (leptokurtic).<br>
-Negative Kurtosis: The data has light tails and a flatter peak (platykurtic).<br>
-Zero Kurtosis: The distribution has a similar tail to a normal distribution (mesokurtic).

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from scipy.stats import skew, kurtosis
```

```python
df = pd.read_csv('US_Top_Ports_Tonnage.csv')
print(df.head(20))
```

```python
skewness = skew(df["Total Tonnage (in million tons)"])
kurt = kurtosis(df["Total Tonnage (in million tons)"])
print("Skewness:",skewness:.2f)
print("Kurtosis:" ,kurt:.2f)
if skewness > 0:
    print("The distribution is positively skewed (right skew).")
elif skewness < 0:
    print("The distribution is negatively skewed (left skew).")
else:
    print("The distribution is approximately symmetric.")

if kurt > 0:
    print("The distribution has heavy tails (leptokurtic).")
elif kurt < 0:
    print("The distribution has light tails (platykurtic).")
else:
    print("The distribution is mesokurtic, similar to a normal distribution.")
```

```python
plt.figure(figsize=(10, 6))
sns.set(style="whitegrid")
sns.histplot(df["Total Tonnage (in million tons)"], kde=True, color="skyblue", bins=10)
plt.title("Histogram of Total Tonnage with KDE", fontsize=14)
plt.xlabel("Total Tonnage (in million tons)", fontsize=12)
plt.ylabel("Frequency", fontsize=12)
plt.show()
```
