#### **Importing necessary libraries for descriptive analysis**
import pandas as pd <br>

##### ii. Load data into desired format

```python
# Create a dictionary with the data
data = {
    "Country": [
        "Albania", "Bulgaria", "Croatia", "Czech Republic", 
        "Hungary", "Poland", "Romania", "Bosnia/Herzegovina"
    ],
    "Per Capita GDP (U.S. $)": [5700, 10700, 13400, 21900, 17600, 14300, 9100, 5600]
}

# Load the data into a pandas DataFrame
df = pd.DataFrame(data)

# Display the DataFrame
print(df)
```

##### iii. Calculate mean of per capita GDP for European countries

```python
# Calculate the mean of the per capita GDP
mean_gdp = df["Per Capita GDP (U.S. $)"].mean()

print(f"The mean per capita GDP for the European countries is ${mean_gdp:,.2f}.")
```

##### iv. Calculate geometric mean of per capita GDP for European countries

```python
from scipy.stats import gmean

# Calculate the geometric mean of the per capita GDP
geometric_mean_gdp = gmean(df["Per Capita GDP (U.S. $)"])

print(f"The geometric mean per capita GDP for the European countries is ${geometric_mean_gdp:,.2f}.")
```

##### v. Calculate harmonic mean of per capita GDP for European countries

```python
from scipy.stats import hmean

# Calculate the harmonic mean of the per capita GDP
harmonic_mean_gdp = hmean(df["Per Capita GDP (U.S. $)"])

print(f"The harmonic mean per capita GDP for the European countries is ${harmonic_mean_gdp:,.2f}.")
```

##### vi. Calculate median of per capita GDP for European countries

```python
# Calculate the median of the per capita GDP
median_gdp = df["Per Capita GDP (U.S. $)"].median()

print(f"The median per capita GDP for the European countries is ${median_gdp:,.2f}.")

plt.figure(figsize=(8, 6))
plt.hist(df["Per Capita GDP (U.S. $)"], bins=8, color="skyblue", edgecolor="black", alpha=0.7)
plt.axvline(x=median_gdp, color="red", linestyle="--", label=f"Median: ${median_gdp:,.2f}")
plt.title("Histogram of Per Capita GDP for European Countries", fontsize=14)
plt.xlabel("Per Capita GDP (U.S. $)", fontsize=12)
plt.ylabel("Frequency", fontsize=12)
plt.legend()
plt.show()
```

##### vii. What is the standard deviation for per capita GDP

```python
# Calculate the standard deviation of the per capita GDP
std_gdp = df["Per Capita GDP (U.S. $)"].std()

print(f"The standard deviation of the per capita GDP for the European countries is ${std_gdp:,.2f}.")
```

##### viii. What is maximum per capita GDP among these 8 countries

##### ix. What is minimum per capita GDP among these 8 countries

```python
# Calculate the minimum per capita GDP
min_gdp = df["Per Capita GDP (U.S. $)"].min()

print(f"The minimum per capita GDP among the 8 European countries is ${min_gdp:,.2f}.")
```

##### x. Compute the mean and standard deviation for Albania, Bulgaria, Croatia, and Czech Republic.

```python
import pandas as pd
# Filter the DataFrame for the selected countries
data = {
    "Country": [
        "Albania", "Bulgaria", "Croatia", "Czech Republic", 
        "Hungary", "Poland", "Romania", "Bosnia/Herzegovina"
    ],
    "Per Capita GDP (U.S. $)": [5700, 10700, 13400, 21900, 17600, 14300, 9100, 5600]
}

df = pd.DataFrame(data)
selected_countries = ["Albania", "Bulgaria", "Croatia", "Czech Republic"]
filtered_df = df[df["Country"].isin(selected_countries)]

# Calculate the mean and standard deviation for the selected countries
mean_selected_gdp = filtered_df["Per Capita GDP (U.S. $)"].mean()
std_selected_gdp = filtered_df["Per Capita GDP (U.S. $)"].std()

print(f"The mean per capita GDP for Albania, Bulgaria, Croatia, and Czech Republic is ${mean_selected_gdp:,.2f}.")
print(f"The standard deviation of per capita GDP for these countries is ${std_selected_gdp:,.2f}.")
```

##### xi. Compute the mean and standard deviation for Hungary, Poland, Romania, and Bosnia/Herzegovina.

```python
# Filter the DataFrame for the selected countries
selected_countries_2 = ["Hungary", "Poland", "Romania", "Bosnia/Herzegovina"]
filtered_df_2 = df[df["Country"].isin(selected_countries_2)]

# Calculate the mean and standard deviation for the selected countries
mean_selected_gdp_2 = filtered_df_2["Per Capita GDP (U.S. $)"].mean()
std_selected_gdp_2 = filtered_df_2["Per Capita GDP (U.S. $)"].std()

print(f"The mean per capita GDP for Hungary, Poland, Romania, and Bosnia/Herzegovina is ${mean_selected_gdp_2:,.2f}.")
print(f"The standard deviation of per capita GDP for these countries is ${std_selected_gdp_2:,.2f}.")
```

##### xii. Use a coefficient of variation to compare the two standard deviations. Treat the data as population data.

```python
# Calculate the CV for the first group (Albania, Bulgaria, Croatia, Czech Republic)
cv_group_1 = (std_selected_gdp / mean_selected_gdp) * 100

# Calculate the CV for the second group (Hungary, Poland, Romania, Bosnia/Herzegovina)
cv_group_2 = (std_selected_gdp_2 / mean_selected_gdp_2) * 100

print(f"The coefficient of variation for Albania, Bulgaria, Croatia, and Czech Republic is {cv_group_1:.2f}%.")
print(f"The coefficient of variation for Hungary, Poland, Romania, and Bosnia/Herzegovina is {cv_group_2:.2f}%.")
```

##### xiii. What are the different tuple values of measure of position
* Maximum value
* Minimum Value
* Median
* First Quartile 
* Second Quartile
* Third Quartile
* Inter Quartile 

```python
# Calculate the required measures of position
max_value = df["Per Capita GDP (U.S. $)"].max()
min_value = df["Per Capita GDP (U.S. $)"].min()
median_value = df["Per Capita GDP (U.S. $)"].median()
first_quartile = df["Per Capita GDP (U.S. $)"].quantile(0.25)
second_quartile = median_value  # Same as the median
third_quartile = df["Per Capita GDP (U.S. $)"].quantile(0.75)
inter_quartile_range = third_quartile - first_quartile

# Display the tuple values
tuple_values = (max_value, min_value, median_value, first_quartile, second_quartile, third_quartile, inter_quartile_range)

print(f"Maximum Value: ${max_value:,.2f}")
print(f"Minimum Value: ${min_value:,.2f}")
print(f"Median Value: ${median_value:,.2f}")
print(f"First Quartile (Q1): ${first_quartile:,.2f}")
print(f"Second Quartile (Q2 - Median): ${second_quartile:,.2f}")
print(f"Third Quartile (Q3): ${third_quartile:,.2f}")
print(f"Interquartile Range (IQR): ${inter_quartile_range:,.2f}")
```

##### xiv. Calculate Z-Score for given data

```python
# Calculate the mean and standard deviation of the per capita GDP
mean_gdp = df["Per Capita GDP (U.S. $)"].mean()
std_gdp = df["Per Capita GDP (U.S. $)"].std()

# Calculate the Z-score for each country
df["Z-Score"] = (df["Per Capita GDP (U.S. $)"] - mean_gdp) / std_gdp

# Display the data with Z-scores
print(df[["Country", "Per Capita GDP (U.S. $)", "Z-Score"]])
```
