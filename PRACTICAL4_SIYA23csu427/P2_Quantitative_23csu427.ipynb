# 1.)<u>`HISTOGRAM`<u>
#### A histogram is a type of chart that 'shows the frequency distribution of data points across a continuous range of numerical values

```python
# Step 1: Import necessary libraries
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Step 2: Load the CSV file
df = pd.read_csv('product.csv')  # Make sure the file path is correct

# Step 3: Create a Histogram for the Sales Amount without the line
plt.figure(figsize=(10, 6))
sns.histplot(df['SALES_AMOUNT'], kde=False, bins=10, color='green', edgecolor='black')

# Step 4: Customize the chart
plt.title('Histogram of Sales Amount', fontsize=16, fontweight='bold')  # Bold title
plt.xlabel('Sales Amount ($)', fontsize=14, fontweight='bold')  # Bold x-axis label
plt.ylabel('Frequency', fontsize=14, fontweight='bold')  # Bold y-axis label

# Step 5: Show the plot
plt.show()
```

# 2)<u>`Frequency Polygon`<u>
#### A frequency polygon is a line graph that shows how often each category of data occurs

```python
# Import necessary libraries
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Load the dataset
df = pd.read_csv('product.csv')  # Ensure the correct path

# Create the histogram and store the output in a variable
plt.figure(figsize=(10, 6))

# Create a histogram with Seaborn, but without the line
hist = sns.histplot(df['SALES_AMOUNT'], kde=False, color='cyan', bins=10, edgecolor='black')

# Get the bin edges and frequencies
bin_edges = hist.patches
bin_centers = [patch.get_x() + patch.get_width() / 2 for patch in bin_edges]
frequencies = [patch.get_height() for patch in bin_edges]

# Add the line (Frequency Polygon) by connecting the bin centers
plt.plot(bin_centers, frequencies, color='green', lw=2, marker='o', linestyle='-', label='Frequency Polygon')

# Customize the plot
plt.title('Histogram with Frequency Polygon of Sales Amount', fontsize=16, fontweight='bold')
plt.xlabel('Sales Amount ($)', fontsize=14, fontweight='bold')
plt.ylabel('Frequency', fontsize=14, fontweight='bold')

# Show the plot
plt.legend()
plt.show()
```

# 3)<u>`OGIVE PLOT`<u>

### An ogive plot, also known as a cumulative frequency polygon, is a graph that shows the cumulative frequency of a set of data.

```python
# Step 1: Import necessary libraries
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np

# Step 2: Load the CSV file
df = pd.read_csv('product.csv')  # Ensure correct file path

# Step 3: Define the number of bins and compute frequency distribution
num_bins = 10
bin_edges = np.histogram_bin_edges(df['SALES_AMOUNT'], bins=num_bins)
freq, edges = np.histogram(df['SALES_AMOUNT'], bins=bin_edges)

# Step 4: Compute cumulative frequency
cumulative_freq = np.cumsum(freq)

# Step 5: Plot the Ogive (Cumulative Frequency Graph)
plt.figure(figsize=(10, 6))
plt.plot(edges[1:], cumulative_freq, marker='o', linestyle='-', color='purple', label='Cumulative Frequency')

# Step 6: Customize the plot
plt.title('Ogive (Cumulative Frequency Plot) of Sales Amount', fontsize=16, fontweight='bold')
plt.xlabel('Sales Amount ($)', fontsize=14, fontweight='bold')
plt.ylabel('Cumulative Frequency', fontsize=14, fontweight='bold')
plt.grid(True)
plt.legend()

# Step 7: Show the plot
plt.show()
```

# 4) <u>`BOX PLOT`<u>

### Box Plot is a graphical method to visualize data distribution for gaining insights and making informed decisions. 
### Box plot is a type of chart that depicts a group of numerical data through their quartiles.

```python
# Step 1: Import necessary libraries
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Step 2: Load the CSV file
df = pd.read_csv('product.csv')  # Ensure the correct file path

# Step 3: Create the Box Plot for the 'SALES_AMOUNT' column
plt.figure(figsize=(8, 6))
sns.boxplot(y=df['SALES_AMOUNT'], color='skyblue')

# Step 4: Customize the plot
plt.title('Box Plot of Sales Amount', fontsize=16, fontweight='bold')  # Bold title
plt.ylabel('Sales Amount ($)', fontsize=14, fontweight='bold')  # Bold y-axis label

# Step 5: Show the plot
plt.show()
```

# 5) <u>`DOT PLOT`<u>

### A dot plot is a graph, similar to a bar graph, that uses dots to represent the number of data points along a number line

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Load dataset
df = pd.read_csv('product.csv')

# Create dot plot for Sales Amount
plt.figure(figsize=(8, 6))
sns.stripplot(y=df['SALES_AMOUNT'], color='red', size=8, jitter=True)

# Customize the plot
plt.title('Dot Plot of Sales Amount', fontsize=16, fontweight='bold')
plt.xlabel('Data Points', fontsize=14, fontweight='bold')  # Manually add X-axis label
plt.ylabel('Sales Amount ($)', fontsize=14, fontweight='bold')

# Show the plot
plt.show()
```

# 7) <u>`Scatter Plot`<u>

### A scatter plot (aka scatter chart, scatter graph) uses dots to represent values for two different numeric variables

```python
# Step 1: Import necessary libraries
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Step 2: Load the dataset
df = pd.read_csv('product.csv')  # Ensure the file path is correct

# Step 3: Create the Scatter Plot
plt.figure(figsize=(8, 6))
sns.scatterplot(x=df['SALES_AMOUNT'], y=df['PROFIT'], color='yellow', edgecolor='black')

# Step 4: Customize the plot
plt.title('Scatter Plot of Sales Amount vs Profit', fontsize=16, fontweight='bold')
plt.xlabel('Sales Amount ($)', fontsize=14, fontweight='bold')
plt.ylabel('Profit ($)', fontsize=14, fontweight='bold')

# Step 5: Show the plot
plt.show()
```
