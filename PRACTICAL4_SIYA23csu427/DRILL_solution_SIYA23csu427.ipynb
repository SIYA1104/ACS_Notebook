### 5)Create Histogram for "Duration" fields in datasets

```python
plt.figure(figsize=(10, 6))
plt.hist(geyser_data['duration'], bins=20, color='skyblue', edgecolor='black')
plt.title('Histogram of Duration')
plt.xlabel('Duration')
plt.ylabel('Frequency')
plt.grid(True)
plt.show()
```

### 6)Create Histogram for "waiting" fileds in datasets

```python
plt.figure(figsize=(10, 6))
plt.hist(geyser_data['waiting'], bins=20, color='lightgreen', edgecolor='black')
plt.title('Histogram of Waiting')
plt.xlabel('Waiting (minutes)')
plt.ylabel('Frequency')
plt.grid(True)
plt.show()
```

### 7)Show the distribution of data along with Histogram
*Duration
*Waiting

```python
# Step 5: Show Distribution for "Duration" and "Waiting" with Histogram and KDE

# Set up the plot for Duration
plt.figure(figsize=(14, 6))

# Plot for Duration
plt.subplot(1, 2, 1)
sns.histplot(geyser_data['duration'], bins=20, kde=True, color='skyblue', edgecolor='black')
plt.title('Distribution of Duration')
plt.xlabel('Duration')
plt.ylabel('Frequency')

# Plot for Waiting
plt.subplot(1, 2, 2)
sns.histplot(geyser_data['waiting'], bins=20, kde=True, color='lightgreen', edgecolor='black')
plt.title('Distribution of Waiting')
plt.xlabel('Waiting (minutes)')
plt.ylabel('Frequency')

# Show the plots
plt.tight_layout()
plt.show()
```

### 8)Plot a histogram of both the total bill and the tips

```python
# Step 1: Import Libraries
import seaborn as sns
import matplotlib.pyplot as plt

# Step 2: Load the "tips" Dataset
tips_data = sns.load_dataset('tips')

# Step 3: Plot Histogram for Total Bill and Tips

# Set up the plot with two subplots
plt.figure(figsize=(14, 6))

# Plot for Total Bill
plt.subplot(1, 2, 1)
sns.histplot(tips_data['total_bill'], bins=20, kde=False, color='lightcoral', edgecolor='black')
plt.title('Histogram of Total Bill')
plt.xlabel('Total Bill ($)')
plt.ylabel('Frequency')

# Plot for Tips
plt.subplot(1, 2, 2)
sns.histplot(tips_data['tip'], bins=20, kde=False, color='lightblue', edgecolor='black')
plt.title('Histogram of Tips')
plt.xlabel('Tip ($)')
plt.ylabel('Frequency')

# Show the plots
plt.tight_layout()
plt.show()
```

### 9)Box plot

```python
# Step 1: Import Libraries
import seaborn as sns
import matplotlib.pyplot as plt

# Step 2: Load the "tips" Dataset
tips_data = sns.load_dataset('tips')

# Step 3: Visualize the Box Plot for Total Bill and Tips

# Set up the plot with two subplots
plt.figure(figsize=(14, 6))

# Box plot for Total Bill
plt.subplot(1, 2, 1)
sns.boxplot(x=tips_data['total_bill'], color='lightcoral')
plt.title('Box Plot of Total Bill')
plt.xlabel('Total Bill ($)')

# Box plot for Tips
plt.subplot(1, 2, 2)
sns.boxplot(x=tips_data['tip'], color='lightblue')
plt.title('Box Plot of Tips')
plt.xlabel('Tip ($)')

# Show the plots
plt.tight_layout()
plt.show()
```

### 10. Creating Histograms and Boxplots Plotted by Groups 
Note: Use Kind  kind  attribute to make groups

```python
# Step 1: Import Libraries
import seaborn as sns
import matplotlib.pyplot as plt

# Step 2: Load the "tips" Dataset
tips_data = sns.load_dataset('tips')

# Step 3: Create Histograms Plotted by Groups
plt.figure(figsize=(14, 6))

# Histogram for Total Bill by Sex
plt.subplot(1, 2, 1)
sns.histplot(data=tips_data, x='total_bill', hue='sex', bins=20, kde=False, palette='pastel', multiple='stack', edgecolor='black')
plt.title('Histogram of Total Bill by Sex')
plt.xlabel('Total Bill ($)')
plt.ylabel('Frequency')

# Boxplot for Total Bill by Time
plt.subplot(1, 2, 2)
sns.boxplot(x='time', y='total_bill', data=tips_data, hue='time', palette='pastel', showfliers=False)  # Use `hue='time'`
plt.title('Box Plot of Total Bill by Time')
plt.xlabel('Time')
plt.ylabel('Total Bill ($)')

# Show the plots
plt.tight_layout()
plt.show()
```
