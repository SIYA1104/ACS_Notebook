# 1) Measure of Central Tendency

### Importing libraries:

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('Sample.csv')

# Display the first 10 rows to verify the data
print(df.head(10))
```

### 1.1 MEAN:

```python
mean_age = df["Age"].mean()
print(mean_age)
```

### 1.2 MEDIAN:

```python
median_age = df["Age"].median()
print(median_age)
```

### 1.3 MODE:

```python
mode_age = df["Age"].mode()[0]
print(mode_age)
```

### 1.4 PERCENTILE & QUARTILE:

```python
# Calculate Percentiles and Quartiles
percentile_25 = np.percentile(df["Age"], 25)
percentile_50 = np.percentile(df["Age"], 50)  # same as median
percentile_75 = np.percentile(df["Age"], 75)
first_quartile = percentile_25
second_quartile = percentile_50
third_quartile = percentile_75
print(f"25th Percentile (Q1): {percentile_25:.2f}")
print(f"50th Percentile (Median): {percentile_50:.2f}")
print(f"75th Percentile (Q3): {percentile_75:.2f}")
print(f"Interquartile Range (IQR): {third_quartile - first_quartile:.2f}")
```

# VISUAL REPRESENTATION:

```python
# Visualization of Central Tendency and Quartiles
plt.figure(figsize=(8, 6))

# Create a boxplot to show the quartiles and the spread of data
sns.boxplot(data=df, y="Age", color="skyblue")

# Add lines for mean, median, and quartiles
plt.axhline(mean_age, color="red", linestyle="--", label=f"Mean: {mean_age:.2f}")
plt.axhline(median_age, color="green", linestyle="--", label=f"Median: {median_age}")
plt.axhline(first_quartile, color="purple", linestyle="--", label=f"Q1: {first_quartile:.2f}")
plt.axhline(third_quartile, color="orange", linestyle="--", label=f"Q3: {third_quartile:.2f}")

# Add titles and labels
plt.title("Measures of Central Tendency & Quartiles for Age", fontsize=14)
plt.ylabel("Age", fontsize=12)
plt.legend()

# Show the plot
plt.show()
```

# 2) Measure of Variablility

### Importing libraries:

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('Sample.csv')

# Display the first 10 rows to verify the data
print(df.head(10))
```

### 2.1 MEAN ABSOLUTE DEVIATION:

```python
# Calculate Measures of Variability

# Mean
mean_age = df["Age"].mean()

# Standard Deviation
std_dev_age = df["Age"].std()

# Mean Absolute Deviation (MAD)
mad_age = df["Age"].apply(lambda x: abs(x - mean_age)).mean()

# Display the results
print(f"Mean: {mean_age:.2f}")
print(f"Standard Deviation: {std_dev_age:.2f}")
print(f"Mean Absolute Deviation (MAD): {mad_age:.2f}")
```

### 2.2 STANDARD DEVIATION:

```python
# Calculate Standard Deviation
std_deviation = df["Age"].std()

# Display the standard deviation
print(f"Standard Deviation of Age: {std_deviation:.2f}")
# Visualization of the standard deviation
plt.figure(figsize=(8, 6))

# Create a histogram to show the distribution of Age
sns.histplot(df["Age"], kde=True, color="skyblue", bins=5, edgecolor="black")

# Add a line for mean
mean_age = df["Age"].mean()
plt.axvline(mean_age, color="red", linestyle="--", label=f"Mean: {mean_age:.2f}")
# Add a line for one standard deviation from the mean
plt.axvline(mean_age + std_deviation, color="green", linestyle="--", label=f"Mean + 1 STD: {mean_age + std_deviation:.2f}")
plt.axvline(mean_age - std_deviation, color="green", linestyle="--", label=f"Mean - 1 STD: {mean_age - std_deviation:.2f}")

# Add titles and labels
plt.title("Distribution of Age with Standard Deviation", fontsize=14)
plt.xlabel("Age", fontsize=12)
plt.ylabel("Frequency", fontsize=12)
plt.legend()

# Show the plot
plt.show()
```

### 2.3 Variance

```python
# Calculate the Mean (already calculated earlier)
mean_age = df["Age"].mean()

# Calculate the Variance (Population Variance, ddof=0)
variance_age = df["Age"].var(ddof=0)

# Calculate the Standard Deviation (Population Standard Deviation)
std_dev_age = df["Age"].std(ddof=0)

# Calculate the Range (Difference between max and min)
range_age = df["Age"].max() - df["Age"].min()

# Display the results
print(f"Mean: {mean_age:.2f}")
print(f"Variance: {variance_age:.2f}")
print(f"Standard Deviation: {std_dev_age:.2f}")
print(f"Range: {range_age:.2f}")

# Visualization of Variability Measures
plt.figure(figsize=(8, 6))

# Create a boxplot to show the spread of data
sns.boxplot(data=df, y="Age", color="skyblue")

# Add lines for mean, standard deviation
plt.axhline(mean_age, color="red", linestyle="--", label=f"Mean: {mean_age:.2f}")
plt.axhline(mean_age + std_dev_age, color="green", linestyle="--", label=f"Mean + 1 SD: {mean_age + std_dev_age:.2f}")
plt.axhline(mean_age - std_dev_age, color="green", linestyle="--", label=f"Mean - 1 SD: {mean_age - std_dev_age:.2f}")

# Add titles and labels
plt.title("Variability Measures: Mean, Standard Deviation, Range", fontsize=14)
plt.ylabel("Age", fontsize=12)
plt.legend()

# Show the plot
plt.show()
```
